```python
def gender(x):
    x = x.replace('  ', ' ')
    x = x.split(' , ')
    return str(x[0][0])

def age(x):
    x = x.replace('  ', ' ')
    return int(x.split(' , ')[1].split(' ')[0])

hh_data['Пол'] = hh_data['Пол, возраст'].apply(gender)
hh_data['Возраст'] = hh_data['Пол, возраст'].apply(age)
```
